$\underline{Question\, 1}$

$\textbf{Equations de Maxwell} :$
$$
\left\{
        \begin{array}{ll}
                          \vec{\nabla}\times\vec{E}=-\partial_t \vec{B} &(1)\\
                          \vec{\nabla}\times\vec{B}=\mu \vec{J}  &(2)\\
                          \vec{\nabla}\cdot\vec{B}=0 &(3)
        \end{array}
\right.
$$


$\textbf{Loi d'Ohm} :\quad  \quad \vec{J}=\sigma \vec{E} \quad (4)$


Prenons le rotationnel de l'équation $(1)$ : 

$\bullet \, \vec{\nabla}\times(\vec{\nabla}\times\vec{B})=\epsilon_{ijk}\partial_j(\epsilon_{klm}\partial_lB_m)=\epsilon_{ijk}\epsilon_{klm}\partial_j\partial_lB_m=(\delta_{il}\delta_{jm}-\delta_{im}\delta_{jl})\partial_j\partial_lB_m=\partial_i\partial_jB_j-\partial_j\partial_jB_i=\vec{\nabla}(\vec{\nabla}\cdot\vec{B})-\Delta\vec{B}\overset{(3)}{=}-\Delta \vec{B}$

$\bullet \, \mu \vec{\nabla}\times\vec{J}\overset{(4)}{=} \mu\sigma\vec{\nabla}\times\vec{E} \overset{(1)}{=}-\mu\sigma\partial_t\vec{B}$

On obtient donc, 

$$\boxed{ \partial_t\vec{B}=\frac{1}{\mu\sigma}\Delta\vec{B}}\quad \quad (\star)$$

qui est bien une équation de diffusion pour le champ magnétique .

$\underline{Question \,2}$

$\textbf{Relations de continuité du champ magnétique à la traversée d'une surface}$

$\rightarrow \,$ Par la conservation du flux du champ magnétique à travers toute surface fermée, on montre que la composante normale du champ magnétique est continue à la traversée d'une surface, ce qui s'écrit :

$$ (\vec{B}_2-\vec{B}_1)\cdot \vec{n}_{21} = 0\quad (5)$$

$\rightarrow \,$ Par le théorème d'Ampère, on peut montrer que la composante tangentielle du champ magnétique est discontinue à la traversée d'une surface, ce qui s'écrit : 

$$ \vec{n}_{21}\times(\vec{B}_2-\vec{B}_1)=\vec{J}\cdot\mu \quad(6)$$

$\vec{n}_{21}$  est le vecteur normal à la surface pointant du milieu 1 ( intérieur du conducteur ) vers le milieu 2 ( extérieur du conducteur ). 


$\quad \quad \quad\textbf{(i) Conditions sur le bord externe du conducteur}$

$\bullet \, \underline{bord \,\, supérieur} : \vec{n}_{21} = \vec{u_y}\quad \Rightarrow  \quad (6)\, : $

$$
\left\{
        \begin{array}{ll}
                          B_{2,z}-B_{1,z}=\mu J_x \\
                          B_{1,x}-B_{2,x}=\mu J_z
        \end{array}
\right.
$$

La composante $z$ du champ magnétique est nulle à l'extérieure du conducteur et le courant sur le bord supérieur est donné par $ \vec{J} = J_x \vec{u_x} $ . 

La deuxième équation donne la continuité de la composante $x$ et l'on obtient : 

Condition au bord supérieur : $$\boxed{ B_z = -\mu J_x}$$

$\bullet \, \underline{bord \,\, inférieur} : \vec{n}_{21} = -\vec{u_y}\quad$  donc par le même argument, avec   $ \vec{J} = -J_x \vec{u_x}$

Condition au bord inférieur : $$\boxed{ B_z = -\mu J_x}$$

$\bullet \, \underline{bord \,\, latéral \,\, gauche} : \vec{n}_{21}=-\vec{u_x} \quad \Rightarrow \quad (6)\, : $
$$
\left\{
        \begin{array}{ll}
                         B_{2,y}-B_{1,y}=\mu J_z\\
                         B_{1,z}-B_{2,z}=-\mu J_y
        \end{array}
\right.
$$

La composante $z$ du champ magnétique est nulle à l'extérieure du conducteur et le courant sur le bord supérieur est donné par $ \vec{J} = J_y \vec{u_y} $ .

La première équation donne la continuité de la composante $y$ et l'on obtient :

Condition sur le bord latéral gauche : $$\boxed{B_z=-\mu J_y}$$

$\bullet \, \underline{bord \,\, latéral \,\, droit} : \vec{n}_{21}=\vec{u_x}\quad $ donc par le même argument, avec 
$ \vec{J} = -J_y \vec{u_y} $   

Condition sur le bord latéral droit : $$\boxed{B_z=-\mu J_y}$$

On trouve finalement une condition générale sur le bord externe du conducteur donnée par

$${\boxed{B_z=-\mu}}$$

ce qui est consistant avec l'intuition et la règle de la main droite qui indique un champ magnétique orienté antiparallèlement à l'axe $\vec{u_z}$


$\quad \quad \quad\textbf{(ii) Conditions sur le bord du supraconducteur}$


$\textbf{Relations de continuité du champ électrique à la traversée d'une surface}$

On a continuité de la composante tangentielle et discontinuité de la composante normale : 

$$
\left\{
       \begin{array}{ll}
                         \vec{n}_{21} \times (\vec{E_2}-\vec{E_1})=0\\
                         \vec{n}_{21} \cdot (\vec{E_2}-\vec{E_1}) =\frac{\sigma_s}{\epsilon}
       \end{array}
\right.
$$

avec $\sigma_s$ la densité surfacique de charges.

Par la loi d'Ohm dont on suppose qu'elle s'applique dans tout le milieu, on déduit facilement que le champ électrique est nul dans la région supraconductrice. 

On en déduit donc que $$\vec{E}=\vec{0}\quad \overset{(4)}{\Leftrightarrow} \quad\vec{J}=\vec{0}\quad \overset{(2)}{\Leftrightarrow}\quad \vec{\nabla}\times\vec{B}=\vec{0}$$

On obtient donc le système d'équations :

$$
\left\{
        \begin{array}{ll}
                         \partial_xB_y-\partial_yB_x = 0\\
                         \partial_yB_z-\partial_zB_y = 0\\
                         \partial_zB_x-\partial_xB_z = 0
        \end{array}
\right.
$$

qui peut être réduit, considérant que les composantes $x$ et $y$ du champ magnétique sont nulles. Par les symétries du problème, le champ magnétique est dirigé suivant $-\vec{u_z}$ .

On obtient finalement les conditions de Neumann aux bords de la cavité supraconductrice : 

$$
\left\{
        \begin{array}{ll}
                      \frac{\partial B_z}{\partial y} = 0 \quad \text{, pour les bords inférieur et supérieur}\\
                      \frac{\partial B_z}{\partial x} = 0\quad \text{, pour les bords latéraux } 
       \end{array}
\right.
$$

$\underline{Question \,3}$

Nous allons calculer l'évolution du champ magnétique au cours du temps dans le domaine par la méthode d'Euler (forward) dans le temps, et avec une discrétisation centrée d'ordre 2 pour le laplacien.

Nous importons les librairies nécessaires.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from matplotlib import animation
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

Introduisons les données initiales du problème.

```python
n=121                               #Nombre de points utilisés dans chaque direction pour discrétiser le domaine
x = numpy.linspace(0,15*10**-3,n)   #Création de l'axe x
y = numpy.linspace(0,15*10**-3,n)   #Création de l'axe y

D=(15*10**(-3))/(n-1)               #Distance entre chaque paire de points (m)

sig=5.96*10**7                      #Conductivité du matériau (S/m)
mu=1.26*10**(-6)                    #Perméabilité magnétique du conducteur (cuivre)
M=numpy.zeros((n,n))                #Matrice du champ magnétique en chaque point du domaine discrétisé
dt=2.5*10**(-7)                     #Valeur du pas de temps pour l'évolution du champ magnétique (s)
T=1.25*10**(-2)                     #Temps maximal utilisé dans l'analyse de l'évolution du champ magnétique (s)
nt=int(T/dt)                        #Nombre de pas de temps maximum utilisé
ncdt=int((5*10**-3)/D)              #Nombre de pas de distance pour atteindre 5*10^-3 m
print(1/(sig*mu)*1/(D**2)*dt, "doit être strictement inférieur à 1/4 pour que l'équation de diffusion soit stable.") 
for i in range (3*ncdt+1):         #Conditions aux bords du conducteur en t=0
            M[0][i]= -mu
            M[3*ncdt][i]=-mu
            M[i][0]=-mu
            M[i][3*ncdt]=-mu 

ims=[]                             #Liste regroupant les images de l'évolution du champs magnétique au cours du temps 
```

Créons une fonction reprenant l'évolution du champ magnétique dans le domaine pour un pas de temps.

Nous utilisons l'équation ci dessus pour avoir notre évolution (forward) dans le temps discrétisée, et centrée dans l'espace.

\begin{equation}
B^{n+1}_{i,j}= B^n_{i,j} + \alpha \left( \frac{dt}{D^2} (B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}) + \frac{dt}{D^2} (B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1})\right) ,
\end{equation} 

Où i et j sont respectivement les indices selon x et y, B est le champ magnétique (T), dt est la valeur du pas de temps (s), D est la distance séparant chaque point (m) et alpha est \begin{equation} \alpha\overset{(\star)}{=} \frac{1}{\mu*\sigma} \end{equation} 

Les condtions aux bords sont de Neumann et sont imposées par:

\begin{equation}
B_{i,bord} = dB\cdot D + B_{i,bord-1}
\end{equation}
ou
\begin{equation}
B_{i,bord} = dB\cdot D + B_{i,bord+1}
\end{equation}
en fonction du bord droit ou gauche.

Et:

\begin{equation}
B_{bord,j} = dB\cdot D + B_{bord-1,j}
\end{equation}
ou
\begin{equation}
B_{bord,j} = dB\cdot D + B_{bord+1,j}
\end{equation}
en fonction du bord du bas ou du haut.

Où dB est la variation de B selon une direction.

Les conditions de Dirichlet sont imposées en donnant assignant directement une valeur au bord.

Le champ magnétique dans le supraconducteur est nul.

```python
def MFED(ncdt,mu,sig,t):
    for i in range (t):
        Mn=M.copy()          #Nous prenons une copie de M pour avoir les valeurs du champ magnétique en chaque point
    
    
       #méthode d'euler dans tout le domaine:

        M[1:-1,1:-1]= Mn[1:-1,1:-1] + (1/(mu*sig))*(dt/(D)**2)*(Mn[2:,1:-1]\
                                -2*Mn[1:-1,1:-1]+ Mn[0:-2,1:-1]+ Mn[1:-1,2:]\
                                -2*Mn[1:-1,1:-1]+ Mn[1:-1,0:-2])
    
        # On replace la bonne valeur du champ dans la région supraconductrice à savoir zéro :
    
        M[ncdt+1,ncdt+1:2*ncdt] =numpy.zeros(ncdt-1)
        M[2*ncdt-1,ncdt+1:2*ncdt]=numpy.zeros(ncdt-1)
        M[ncdt+1:2*ncdt,ncdt+1]=numpy.zeros(ncdt-1)
        M[ncdt+1:2*ncdt,2*ncdt-1]=numpy.zeros(ncdt-1)
   
    
        #Nous imposons les conditions aux bords du supraconducteur
        M[ncdt,ncdt:2*ncdt+1]=M[ncdt-1,ncdt:2*ncdt+1] 
        M[2*ncdt,ncdt:2*ncdt+1]=M[2*ncdt+1,ncdt:2*ncdt+1] 
        M[ncdt:2*ncdt+1,ncdt]=M[ncdt:2*ncdt+1,ncdt-1] 
        M[ncdt:2*ncdt+1,2*ncdt]=M[ncdt:2*ncdt+1,2*ncdt+1]
    
        #Nous imposons les conditions aux bords du conducteur
        for i in range (3*ncdt+1):
            M[0][i]= -mu
            M[3*ncdt][i]=-mu
            M[i][0]=-mu
            M[i][3*ncdt]=-mu 
    
    return (M)
```

Vérifions que la fonction fonctionne bien en l'utilisant nt fois, pour arriver à un temps de 1.25*10^-2 s.

```python
MFED(ncdt,mu,sig,nt)
```

Nous regardons le graphique reprenant le champ magnétique (en T) en chaque point en t=1.25*10^-2 s.

```python
pyplot.figure(figsize=(9,7))                    #Type de graphique donné par les Jupyter Notebooks
pyplot.contourf(y,x,M,100,cmap=cm.viridis)
pyplot.ylabel('Position y en m')
pyplot.xlabel('Position x en m')
pyplot.colorbar();
```

$\underline{Question \,6}$

Que vaut, à l'instant $t=1.25\cdot 10^{-3}s$, la valeur de la densité de courant éléctrique ${\bf J}$ en $(x,y,z)=(2.5\cdot 10^{-3}{\rm m},2.5\cdot 10^{-3}{\rm m},0 {\rm m})$?

Nous savons que $$\vec{\nabla}\times\vec{B}=\mu \cdot \vec{J}$$

et donc que \begin{array}{ll}
                          \mu \cdot J_x= \partial_y B_z \\
                          \mu \cdot J_y=-\partial_x B_z  \\
                          \mu \cdot J_z=0
            \end{array}
car les composantes du champ magnétique sont nulles selon x et y.

Où $J_x$ est la composante du vecteur densité de courant $\vec{J}$ selon x, $J_y$ est la composante selon y et $J_z$ est la composante selon z.

Nous approximons les dérivées partielles pour avoir la valeur de chaque composante de $\vec{J}$.

```python
M=numpy.zeros((n,n))      #On réinitialise la matrice M
ntn=int(1.25*10**-3/dt)   #nombre d'ittération
nxy=int(2.5*10**-3/D)     #Indice de position
MFED(ncdt,mu,sig,ntn)     #Nous regardons l'évolution du champs magnétique après 1.25*10^-3s
Bxy=M[nxy][nxy]           #La valeur du champ magnétique en la position demandée
Bx_y=M[nxy][nxy-1]        #La valeur du champ magnétique en x = 2.5*10e-3 - dx mm y = 2.5*10e-3mm et z=0
Bxy_=M[nxy-1][nxy]        #La valeur du champ magnétique en x = 2.5*10e-3 mm y = 2.5*10e-3 - dy mm et z=0
Jx=(Bxy-Bx_y)/(mu*D)      #La composante x de J
Jy=-(Bxy-Bxy_)/(mu*D)     #La composante y de J
Jz=0                      #La composante z de J
J=[Jx,Jy,Jz]
Jv=numpy.sqrt(Jx**2+Jy**2) #La valeur de J en la position
print(J,'est le vecteur densité de courant et',Jv,'est la valeur de J en A/m²')
```

La valeur de la densité de courant éléctrique ${\bf J}$ en $(x,y,z)=(2.5\cdot 10^{-3}{\rm m},2.5\cdot 10^{-3}{\rm m},0 {\rm m})$ est  $31.686 A/m²$

$\underline{Question\,7}$

Que vaut, à l'instant $t=1.25\cdot 10^{-2}s$, la valeur du flux du champ magnétique à travers la section $z=0$ du conducteur? Justifier cette valeur.

Nous allons sommer les valeurs pour le champ magnétique en chaque point de M.

```python
M=numpy.zeros((n,n))          #On réinitialise la matrice M
FTA=0                         #FTA va représenter la somme de chaque valeur du champ magnétique en chaque point
MFED(ncdt,mu,sig,nt)          #Nous regardons l'évolution du champs magnétique après 1.25*10^-2s
for j in range (n):           #Nous effectuons la somme en question
    for k in range (n):
        FTA=M[j][k]+FTA
        
A= 15*10**(-3)*15*10**(-3)    #Aire de la section du conducteur
print (FTA*(A), 'T*m² est le flux magnétique à travers la section du conducteur en z=0')
            
```

Le flux magnétique est cohérent avec l'aire de la section du conducteur et les courants surfaciques de 1A. Il est négatif à cause de l'influence négative de chaque bord du conducteur sur le champ magnétique.

Après $1.25\cdot 10^{-2}s$, le flux est devenu quasi-maximal car le champ magnétique se stabilise en une valeur maximale en chaque point. Le champ magnétique étant presque stabilisé, la densité de courant devient presque nulle.
